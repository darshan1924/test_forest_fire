## Data cleaning

```python
# Missing values
dataset[dataset.isnull().any(axis=1)]
```

The dataset is converted into two sets based on region from 122th index, we can make a new column based on the region

1 : 'Bejala Region Dataset'
2 : 'Sidi-bel abbes region dataset'

Add new column with region 

```python
dataset.loc[:122, "Region"] = 0
dataset.loc[122: , "Region"] = 1
df = dataset
```

```python
df.info()
```

```python
df[['Region']] = df[['Region']].astype(int)
```

```python
df.isnull().sum()
```

```python
# Remove the null values 
df = df.dropna().reset_index(drop=True)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.iloc[[122]]
```

```python
df = df.drop(122).reset_index(drop=True)
```

```python
df.columns
```

```python
# fix the spaces in column name
df.column = df.columns.str.strip()
```

```python
df.column
```

```python
df.head()
```

```python
# Change the required column in interger or float

df[['day', 'month', 'year', 'Temperature']] = df[['day', 'month', 'year', 'Temperature']].astype(int)
```

```python
df.columns = df.columns.str.strip()
df.columns = df.columns.str.replace(r'\s+', '', regex=True)
```

```python
df.columns
```

```python
df[['RH', 'Ws']] = df[['RH', 'Ws']].astype(int)
```

```python
df.info()
```

```python
# Changing the other column to float data type

objects = [features for features in df.columns if df[features].dtype=='O']
```

```python
for i in objects:
    if i!='Classes':
        df[i] = df[i].astype(float)
```

```python
df.info()
```

```python
df.describe()
```

```python
# Lets save the cleaned dataset
df.to_csv('forest_fire_cleaned_dataset.csv', index=False)
```
